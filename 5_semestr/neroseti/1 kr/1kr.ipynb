```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import MinMaxScaler

from sklearn import metrics

datSet = pd.read_csv("bikes_rent.csv",sep=",",header=0)
datSet.shape
datSet.describe()
```

```python
x = datSet[['mnth', 'workingday','temp','hum','windspeed(mph)']].values
y = datSet['cnt'].values

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42) 

min_max_scale=MinMaxScaler()

x_minMax=min_max_scale.fit_transform(x)
x_train_min_max, x_test_min_max, y_train_min_max, y_test_min_max = train_test_split(x_minMax, y, test_size=0.3, random_state=42)

regressor=LinearRegression()
regressor2=LinearRegression()

regressor.fit(x_train,y_train)
regressor2.fit(x_train_min_max,y_train_min_max)

y_pred=regressor.predict(x_test)
y_pred2=regressor.predict(x_train)

y_pred3=regressor2.predict(x_test_min_max)
y_pred4=regressor2.predict(x_train_min_max)
```

```python
coff2= pd.DataFrame(regressor2.coef_,columns=['Coefficient'])
coff2
```

```python
df=pd.DataFrame({"Фактический ": y_test.flatten(),"Прогнозируемый ": y_pred.flatten()})
df
```

```python
df=pd.DataFrame({"Фактический": y_train.flatten(),"Прогнозируемый": y_pred2.flatten()})
df
```

```python
print("Среднекватратичное отклонение тестовые ",metrics.mean_squared_error(y_test,y_pred))
print("Среднекватратичное отклонение тренировочное ",metrics.mean_squared_error(y_train,y_pred2))

print("Среднекватратичное отклонение тестового для максимизируемого ",metrics.mean_squared_error(y_test_min_max,y_pred3))
print("Среднекватратичное отклонение тренировочное для максимизируемого ",metrics.mean_squared_error(y_train_min_max,y_pred4))
```
